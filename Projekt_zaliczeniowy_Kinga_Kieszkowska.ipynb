```python
def osoby(text):
  doc = sp(text)
  ents = doc.ents
  return ' '.join([ent.text for ent in ents if ent.label_ == "persName"])

df["persons"] = df["text"].apply(osoby)
```

Zadanie 2

```python
def miejsca (text):
  doc = sp(text)
  ents = doc.ents
  return ' '.join([ent.text for ent in ents if ent.label_ == "placeName"])

df["places"] = df["text"].apply(miejsca)
```

Zadanie 3

```python
def organizacje (text):
  doc = sp(text)
  ents = doc.ents
  return ' '.join([ent.text for ent in ents if ent.label_ == "orgName"])

df["organizations"] = df["text"].apply(organizacje)
```

```python
import spacy
```

```python
nlp_blank = spacy.blank("pl")
ruler = nlp_blank.add_pipe("entity_ruler")

pattern = "\S+"
patterns = [{"label": "HASHTAG", "pattern": [{"ORTH": "#"}, {"TEXT": {"REGEX": pattern}}]}]

ruler.add_patterns(patterns)

def hashtag(tekst):
  doc = nlp_blank(tekst)
  result = ""
  for ent in doc.ents:
    if ent.label_ == "HASHTAG":
      result += ent.text + " "
  return result

df_tvp["hashtags"] = df_tvp["text"].apply(hashtag)

df_tvp
```
